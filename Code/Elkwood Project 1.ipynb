```python
import pandas as pd
import pandas as pd
import numpy
```

```python
from selenium import webdriver
driver = webdriver.Chrome()
driver.get('https://www.rottentomatoes.com/top/bestofrt/top_100_horror_movies/')
```

To Get the Total Line for Each Movie

```python
total_body= driver.find_elements_by_tag_name('tbody')[2]
rows= total_body.find_elements_by_tag_name('tr')
movies = []

for row in rows:
    cells= (row.find_elements_by_tag_name('td'))
    movie={}
    movie['rank']=(cells[0].text)
    movie['score']=(cells[1].text)
    movie['title']=(cells[2].text)
    movies.append(movie)
    
    
```

```python
total_body.get_attribute('href')
```

```python
df_movies= pd.DataFrame(movies)
```

```python
df_movies.to_csv("df_movies.csv", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_movie_with_bs(row):
    try:
        response = requests.get(row['link'])
        doc = BeautifulSoup(response.text, 'html5lib')
        movie_stats=doc.find_all('ul')[14]
        synopsis=(doc.find(class_="movie_synopsis").text)
        rating=movie_stats.find_all('div')[1].text
        director=movie_stats.find_all('div')[5].text
        writer=movie_stats.find_all('div')[7].text
        box_office=movie_stats.find_all('div')[13].text
        run_time=movie_stats.find_all('div')[15].text
        studio=movie_stats.find_all('div')[17].text

        return pd.Series({
            'synopsis': synopsis,
            'rating': rating,
            'director': director,
            'writer': writer,
            'box_office': box_office,
            'run_time': run_time,
            'studio': studio
        })
    except: 
            print("No")
        
```

```python
bs_df = df_all_movies.apply(get_movie_with_bs, axis=1)
```
